# 2. Speech rate and information density

Hmm... what an intriguing relationship we just revealed... let’s keep these clusters in mind while exploring whether continuing our exploration. The two variables that directly determine the information rate of a language are its information density (bits of information per syllable) and the rate at which it is spoken (syllables per second). A high value in both would indicate a high information rate and efficient communication (a higher number of information bits conveyed per second).

<div class="alert alert-success" style="color:black">

### Question 2
rubric={reasoning:2,accuracy:3}
    
<h4>R</h4>
<ol type="1">
<li>While the information density of each language have already been provided to us, we will have to approximate the speech rate (syllables per second) of each language via the <code>spoken_texts</code> dataset. Using this data, calculate the average speech rate per language (for now, treat all entries as individual observations, even if some are read by the same person).</li>
    <li>Create a new column, <code>speech_rate</code>, in the <code>languages</code> data frame that hold these average values. Since the calculated values are not in the same order as the languages in the <code>languages</code> data frame, you cannot simply append it. Instead, you need to use the <code>inner_join</code> function to merge the dataframe of average speech rate values that you just created with the <code>languages</code> dataframe. The language ISO codes could be used as the column to join on.
<li>Create a scatter plot of the information density versus the speech rate for the <code>languages</code> dataframe. Use the same groupings as in the previous scatter plot you made and add one trend line per group. Label the same points with text labels as in the previous scatterplot.</li>
<li>Write a few sentences under the plot describing the relationship you observe. Based on this plot, do speakers of any particular languages appear to be more efficient at conveying information?</li>
</ol>
    
</div>

```python
# YOUR ANSWER GOES HERE
speech_rate <- spoken %>%
  mutate(speech_rate = syllables / duration) %>%
  group_by(iso_lang) %>%
  dplyr::summarize(speech_rate = mean(speech_rate, na.rm = TRUE), .groups = 'drop')

language <- language %>%
  inner_join(speech_rate, by = "iso_lang")
```

```python
plotB = ggplot(data= language, aes(x=information_density, y=speech_rate, color = continent)) +
  geom_point(size = 3) +
  geom_smooth(method = "lm", se = TRUE) +
  geom_text(aes(label = iso_lang), vjust = -1, size = 3) +
  labs(x="Information Density",
       y="Speech Rate")

print(plotB)
```

YOUR ANSWER TO 4 GOES HERE

There is a pretty strong negative relationship between speech rate and information density, and French and English appear to be above the trend line the most. Therefore French and English are the most efficient at conveying information.

# 3. Language information rate

### Question 3

<h4>R</h4>
    
<ol type="1">
<li>It looks like we do not have a language that is high both in information density and speech rate, which could indicate that human minds are not good at processing auditory information beyond a certain rate limit (read more in the article if you're interested). Let’s directly plot the rate at which information is conveyed for all the languages to see if there are at least small differences between them. Combine the appropriate columns to create a new column, <code>information_rate</code>, which holds the amount of information per second conveyed by each speaker.</li>
<li>Create a plot that has the language ISO codes on one axis and the information rate on the other. Plot a graphical representation of either the distribution of data or every single data point. Color this by the same variable as in the previous two scatter plots. Sort the languages in a suitable order.</li>
<li>Include a marks for the mean and the 95% confidence interval on top of the mark you just created above. Color this in a suitable way so that it stands out from the distribution/individual observations created in the previous point.</li>
</ol>
</div>

```python
# YOUR ANSWER GOES HERE
spoken_with_rate <- spoken %>%
  left_join(language %>% select(iso_lang, information_density, continent), by = "iso_lang") %>%
  mutate(speech_rate = syllables / duration,
         information_rate = information_density * speech_rate)

language <- language %>%
  inner_join(spoken_with_rate %>%
               group_by(iso_lang) %>%
               dplyr::summarize(information_rate = mean(information_rate, na.rm = TRUE),
                         continent = first(continent), .groups = 'drop'),
             by = "iso_lang")
```

```python
summary_stats <- spoken_with_rate %>%
  group_by(iso_lang) %>%
  dplyr::summarize(
    mean_rate = mean(information_rate, na.rm = TRUE),
    n = n(),
    se = sd(information_rate, na.rm = TRUE) / sqrt(n),
    ci_lower = mean_rate - 1.96 * se,
    ci_upper = mean_rate + 1.96 * se,
    continent = first(continent),
    .groups = 'drop'
  )

plotC = ggplot(data = spoken_with_rate, 
       aes(x = reorder(iso_lang, information_rate, FUN = mean), 
           y = information_rate, 
           color = continent)) +
  geom_jitter(width = 0.2, alpha = 0.5, size = 2) +
  geom_point(data = summary_stats, 
             aes(x = reorder(iso_lang, mean_rate), y = mean_rate), 
             color = "black", size = 4, shape = 18, inherit.aes = FALSE) +
  geom_errorbar(data = summary_stats, 
                aes(x = reorder(iso_lang, mean_rate), ymin = ci_lower, ymax = ci_upper),
                color = "black", width = 0.3, size = 1, inherit.aes = FALSE) +
  labs(x = "Language",
       y = "Information Rate (bits/second)",
       color = "Continent") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
print(plotC)
```

```python
# Explore speaker-level variation
# Reload language data to ensure we have continent column
language_base <- read.csv("languages.csv")

# Calculate mean statistics by speaker
speaker_stats <- spoken %>%
  left_join(language_base %>% select(iso_lang, information_density, continent), by = "iso_lang") %>%
  mutate(speech_rate = syllables / duration,
         information_rate = information_density * speech_rate) %>%
  group_by(speaker, iso_lang, sex, age, continent) %>%
  dplyr::summarize(
    mean_speech_rate = mean(speech_rate, na.rm = TRUE),
    mean_information_rate = mean(information_rate, na.rm = TRUE),
    n_texts = n(),
    .groups = 'drop'
  )

# Get language-level statistics for reference
language_speaker_summary <- speaker_stats %>%
  group_by(iso_lang, continent) %>%
  dplyr::summarize(
    mean_speaker_rate = mean(mean_speech_rate, na.rm = TRUE),
    sd_speaker_rate = sd(mean_speech_rate, na.rm = TRUE),
    n_speakers = n(),
    .groups = 'drop'
  )

head(speaker_stats, 10)
head(language_speaker_summary)
```

```python
# Figure 1: Speech rate variation by speaker within each language
plot1 <- ggplot(speaker_stats, aes(x = reorder(iso_lang, mean_speech_rate, FUN = median), 
                                    y = mean_speech_rate,
                                    fill = continent)) +
  geom_boxplot(alpha = 0.7, outlier.alpha = 0.5) +
  geom_jitter(width = 0.15, alpha = 0.4, size = 2, color = "black") +
  labs(title = "Speech Rate Variation: Comparing Individual Speakers Within Languages",
       x = "Language",
       y = "Mean Speech Rate (syllables/second)",
       fill = "Continent") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        legend.position = "right")

print(plot1)
```

```python
# Figure 2: Information rate by speaker within each language
plot2 <- ggplot(speaker_stats, aes(x = reorder(iso_lang, mean_information_rate, FUN = median), 
                                    y = mean_information_rate,
                                    fill = continent)) +
  geom_boxplot(alpha = 0.7, outlier.alpha = 0.5) +
  geom_jitter(width = 0.15, alpha = 0.4, size = 2, color = "black") +
  labs(title = "Information Rate Variation: Individual Speakers",
       x = "Language",
       y = "Mean Information Rate (bits/second)",
       fill = "Continent") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        legend.position = "none")

print(plot2)
```

```python
# Figure 3: Effect of sex and age on speaker variation
# Calculate statistics by sex and language
speaker_stats_with_analysis <- speaker_stats %>%
  group_by(iso_lang, continent) %>%
  mutate(language_mean_rate = mean(mean_speech_rate, na.rm = TRUE),
         deviation_from_language_mean = mean_speech_rate - language_mean_rate) %>%
  ungroup()

# Create summary for comparison
sex_age_summary <- speaker_stats %>%
  group_by(iso_lang, sex, continent) %>%
  dplyr::summarize(
    mean_rate_by_sex = mean(mean_speech_rate, na.rm = TRUE),
    n = n(),
    .groups = 'drop'
  ) %>%
  filter(!is.na(sex))

plot3 <- ggplot(sex_age_summary, aes(x = reorder(iso_lang, mean_rate_by_sex, FUN = mean), 
                                      y = mean_rate_by_sex,
                                      fill = sex)) +
  geom_col(position = "dodge", alpha = 0.8) +
  facet_wrap(~continent, scales = "free_x") +
  labs(title = "Do Males and Females Speak at Different Rates? (By Speaker Average)",
       x = "Language",
       y = "Mean Speech Rate (syllables/second)",
       fill = "Speaker Sex") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 8))

print(plot3)
```

## Story: How Much Do Speakers Vary Within Their Own Language?

In the main analysis, we found that all languages cluster around similar information rates, suggesting a universal cognitive constraint. But this narrative assumes all speakers within a language speak similarly. This raises the question of how much individual variation exists within each language community compared to the differences between languages? I grouped by speaker and calculated their mean speech rate across all texts they read. This approach better reveals individual differences that might be masked when pooling all observations.

### Figure 1: Speaker-Level Speech Rate Variation by Language

The first visualization shows the distribution of speaker-level mean speech rates for each language. Rather than just seeing the overall mean and confidence interval, we can now see the spread of individual speakers. This visualization reveals that while languages do cluster differently, individual speakers within the same language vary considerably. Some speakers are significantly faster or slower than others, suggesting that factors beyond language structure—such as age, personality, or speaking style—influence speech rate.

### Figure 2: Information Rate at the Speaker Level

Since information rate combines both speech rate and information density, examining speaker-level information rates shows whether linguistic constraints really do keep speakers uniform in their actual information transmission. Interestingly, the information rate plot shows less variation than the speech rate plot. This suggests that the linguistic constraint we discovered earlier—that fast-speaking languages have low information density while slow-speaking languages have high information density. Individual speakers naturally modulate their speech rate based on the structure of their language, keeping their information transmission remarkably consistent. A speaker might deviate from their language group's typical speech rate, but they're unconsciously compensating through their language's information density.

### Figure 3: Beyond Language Structure—The Role of Speaker Demographics

But are all speakers within a language truly equivalent? The main analysis pooled all speakers together, but speakers vary in sex and age. Do these demographic factors explain some of the within-language variation we observe? This comparison reveals a consistent pattern. Across nearly every language and continent, male speakers tend to speak slightly faster than female speakers. This demographic effect is small but systematic, and it persists even after controlling for language structure. The consistency of this pattern across diverse linguistic families suggests it reflects biological or social factors that transcend language boundaries.

## Conclusion

By shifting from text-level to speaker-level analysis, we've uncovered a richer picture of human speech variation. While the original analysis showed that all languages achieve similar information rates, we now see that individual speakers introduce considerable variation around these linguistic averages. Demographic factors like sex appear to systematically influence speech rate. Yet, speakers seem to naturally adjust their speech to maintain relatively stable information transmission rates, suggesting that the human cognitive limit on information processing is truly fundamental, not just at the language level, but reflected in the behavior of individual speakers across the world.
