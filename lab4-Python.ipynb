```python
# YOUR ANSWER GOES HERE
import pandas as pd
languages=pd.read_csv('languages.csv')
spoken_texts=pd.read_csv('spoken-texts.csv')
```

```python
languages.head()
```

```python
import numpy as np, pandas as pd, altair as alt
import statsmodels.api as sm

x, y, g = "information_density", "distinct_syllables", "continent"
df = languages.query("iso_lang != 'FRA'")[["language", x, y, g]].dropna()

pred = []
for k, d in df.groupby(g):
    X = sm.add_constant(d[x])
    m = sm.OLS(d[y], X).fit()
    grid = np.linspace(d[x].min(), d[x].max(), 80)
    pr = m.get_prediction(sm.add_constant(grid)).summary_frame(alpha=0.05)
    pred.append(pd.DataFrame({x:grid, g:k, "yhat":pr["mean"], "lo":pr["mean_ci_lower"], "hi":pr["mean_ci_upper"]}))
pred = pd.concat(pred, ignore_index=True)

points = alt.Chart(df).mark_point(size=60, opacity=0.6).encode(
    x=alt.X(f"{x}:Q", scale=alt.Scale(zero=False)),
    y=alt.Y(f"{y}:Q", scale=alt.Scale(zero=False)),
    color=f"{g}:N",
    tooltip=[f"{g}:N", "language:N", f"{x}:Q", f"{y}:Q"]
)

band = alt.Chart(pred).mark_area(opacity=0.25).encode(
    x=f"{x}:Q", y="lo:Q", y2="hi:Q", color=f"{g}:N"
)

line = alt.Chart(pred).mark_line().encode(
    x=f"{x}:Q", y="yhat:Q", color=f"{g}:N"
)

plot_A=points + band + line
plot_A
```

```python
# YOUR ANSWER GOES HERE
spoken_texts
```

```python
speech = (
    spoken_texts
    .assign(speech_rate=spoken_texts["syllables"] / spoken_texts["duration"])
    .groupby("iso_lang", as_index=False)["speech_rate"]
    .mean()
)
languages = languages.merge(speech, on="iso_lang", how="left")
```

```python
languages.head()
```

```python
x, y, g = "information_density", "speech_rate", "continent"
df = languages.query("iso_lang != 'FRA'")[["language", x, y, g]].dropna()

pred = []
for k, d in df.groupby(g):
    X = sm.add_constant(d[x])
    m = sm.OLS(d[y], X).fit()
    grid = np.linspace(d[x].min(), d[x].max(), 80)
    pr = m.get_prediction(sm.add_constant(grid)).summary_frame(alpha=0.05)
    pred.append(pd.DataFrame({x:grid, g:k, "yhat":pr["mean"], "lo":pr["mean_ci_lower"], "hi":pr["mean_ci_upper"]}))
pred = pd.concat(pred, ignore_index=True)

points = alt.Chart(df).mark_point(size=60, opacity=0.6).encode(
    x=alt.X(f"{x}:Q", scale=alt.Scale(zero=False)),
    y=alt.Y(f"{y}:Q", scale=alt.Scale(zero=False)),
    color=f"{g}:N",
    tooltip=[f"{g}:N", "language:N", f"{x}:Q", f"{y}:Q"]
)

band = alt.Chart(pred).mark_area(opacity=0.25).encode(
    x=f"{x}:Q", y="lo:Q", y2="hi:Q", color=f"{g}:N"
)

line = alt.Chart(pred).mark_line().encode(
    x=f"{x}:Q", y="yhat:Q", color=f"{g}:N"
)

plot_B=points + band + line
plot_B
```

YOUR ANSWER TO 4 GOES HERE

We observe a clear negative relationship between information density and speech rate: as information density increases, languages tend to be spoken more slowly. No language appears to be particularly efficient in the sense of simultaneously exhibiting both high information density and a high speech rate. However, English stands out slightly, as it combines relatively high information density while maintaining a comparatively high speech rate.

<li>It looks like we do not have a language that is high both in information density and speech rate, which could indicate that human minds are not good at processing auditory information beyond a certain rate limit (read more in the article if you're interested). Letâ€™s directly plot the rate at which information is conveyed for all the languages to see if there are at least small differences between them. <li>Create a plot that has the language ISO codes on one axis and the information rate on the other. Plot a graphical representation of either the distribution of data or every single data point. Color this by the same variable as in the previous two scatter plots. Sort the languages in a suitable order.</li>

```python
obs = (
    spoken_texts
    .merge(
        languages[["iso_lang", "information_density", "continent"]],
        on="iso_lang",
        how="left"
    )
    .assign(
        speech_rate=lambda d: d["syllables"] / d["duration"],
        information_rate=lambda d: d["information_density"] * d["speech_rate"]
    )
    .dropna(subset=["information_rate"])
)

obs
```

```python
base_plot = alt.Chart(obs).mark_circle(opacity=0.35, size=10).encode(
    y=alt.Y("iso_lang:N", sort="-x", title="Language (ISO)"),
    x=alt.X("information_rate:Q", title="Information rate (info/sec)"),
    color="continent:N",
    tooltip=["iso_lang:N", "information_rate:Q"]
).properties(
    title="Information Rate per Speaker by Language",
    width=800
)

summary = (
    obs
    .groupby(["iso_lang", "continent"], as_index=False)
    .agg(
        mean_rate=("information_rate", "mean"),
        sd_rate=("information_rate", "std"),
        n=("information_rate", "count")
    )
)

summary["se"] = summary["sd_rate"] / np.sqrt(summary["n"])
summary["ci"] = 1.96 * summary["se"]
summary["lo"] = summary["mean_rate"] - summary["ci"]
summary["hi"] = summary["mean_rate"] + summary["ci"]

ci = alt.Chart(summary).mark_errorbar(color="black").encode(
    y="iso_lang:N",
    x="lo:Q",
    x2="hi:Q"
)

mean = alt.Chart(summary).mark_point(color="black", size=20).encode(
    y="iso_lang:N",
    x="mean_rate:Q"
)

plot_C=(base_plot + ci + mean)
plot_C
```

```python
lang_text = (
    obs.groupby(["iso_lang", "continent", "text"], as_index=False)
       .agg(mean_rate=("speech_rate", "mean"))
)

text_mean = lang_text.groupby("text", as_index=False).agg(text_avg=("mean_rate", "mean"))
rel = lang_text.merge(text_mean, on="text")
rel["relative_rate"] = rel["mean_rate"] - rel["text_avg"]

lang_order = (
    rel.groupby("iso_lang", as_index=False)
       .agg(overall_mean=("mean_rate", "mean"))
       .sort_values("overall_mean", ascending=True)  # True = ä»Žæ…¢åˆ°å¿«
)

lang_sort = lang_order["iso_lang"].tolist()

fig_rel_sorted = alt.Chart(rel).mark_rect().encode(
    x=alt.X("text:N", title="Text"),
    y=alt.Y(
        "iso_lang:N",
        title="Language (sorted by overall mean speed)",
        sort=lang_sort
    ),
    color=alt.Color(
        "relative_rate:Q",
        title="Relative speech rate vs text mean",
        scale=alt.Scale(scheme="redblue", domainMid=0)
    ),
    tooltip=[
        alt.Tooltip("iso_lang:N", title="Language"),
        alt.Tooltip("text:N", title="Text"),
        alt.Tooltip("mean_rate:Q", title="Mean rate", format=".2f"),
        alt.Tooltip("relative_rate:Q", title="Relative rate", format=".2f"),
        alt.Tooltip("n:Q", title="n")
    ]
).properties(
    title="Relative speech rate by language within each text (languages sorted)",
    width=720,
    height=420
)

fig_rel_sorted
```

```python
ct = (
    obs.groupby(["continent", "text"], as_index=False)
       .agg(
           mean_rate=("speech_rate", "mean"),
           sd=("speech_rate", "std"),
           n=("speech_rate", "size")
       )
)

cont_order = (
    ct.groupby("continent", as_index=False)
      .agg(overall_mean=("mean_rate", "mean"))
      .sort_values("overall_mean", ascending=True)
)
cont_sort = cont_order["continent"].tolist()
```

```python
fig_cont_line = alt.Chart(ct).mark_line(point=True).encode(
    x=alt.X("text:N", title="Text"),
    y=alt.Y("mean_rate:Q", title="Mean speech rate (syllables/sec)",scale=alt.Scale(zero=False)),
    color=alt.Color("continent:N", sort=cont_sort, title="Continent"),
    tooltip=[
        alt.Tooltip("continent:N", title="Continent"),
        alt.Tooltip("text:N", title="Text"),
        alt.Tooltip("mean_rate:Q", title="Mean rate", format=".2f"),
        alt.Tooltip("n:Q", title="n obs")
    ]
).properties(
    title="Mean speech rate across texts by continent",
    width=720,
    height=380
)

fig_cont_line
```

```python
lang_text_var = (
    lang_text
    .groupby(["iso_lang", "continent"], as_index=False)
    .agg(text_sd=("mean_rate", "std"))
    .sort_values("text_sd", ascending=False)
)


fig_var = alt.Chart(lang_text_var).mark_bar().encode(
    x=alt.X("text_sd:Q", title="SD of mean speech rate across texts"),
    y=alt.Y("iso_lang:N", sort='-x', title="Language"),
    color=alt.Color("continent:N", title="Continent"),
    tooltip=[
        alt.Tooltip("iso_lang:N", title="Language"),
        alt.Tooltip("continent:N", title="Continent"),
        alt.Tooltip("text_sd:Q", format=".2f", title="SD across texts")
    ]
).properties(
    title="Sensitivity of languages to text content",
    width=420,
    height=360
)

fig_var
```

```python
left = fig_rel_sorted.properties(
    width=520,
    height=480
)

right_top = fig_cont_line.properties(
    width=360,
    height=200
)

right_bottom = fig_var.properties(
    width=360,
    height=200
)

final_fig = alt.hconcat(
    left,
    alt.vconcat(
        right_top,
        right_bottom
    )
).resolve_scale(
    color='independent'
).properties(
    title=alt.TitleParams(
        text="Effect of text on relative speech rate across languages and continents",
        anchor="middle",
        fontSize=20
    )
)

final_fig
```

Figure A shows text-normalized relative speech rates by language, revealing that while absolute speaking speed varies across texts, the relative ordering of languages is remarkably stable. Languages that are faster on average tend to remain faster across nearly all texts, and slower languages remain consistently slower, indicating that text content does not substantially alter which languages are faster. Figure B aggregates speech rates at the continent level and shows that Europe consistently exhibits higher mean speech rates than Asia across all texts, with parallel trends suggesting minimal interaction between text and continent. Finally, Figure C quantifies sensitivity to text content by measuring variability in mean speech rate across texts. Although some languages display greater text sensitivity than others, this variability affects the magnitude of speech rate rather than reversing language or continent-level rankings. Taken together, these results suggest that text content influences overall speaking speed but does not fundamentally change relative speech rate patterns across languages or continents.

When you are ready to submit your assignment do the following:

1. Run all cells in your notebook to make sure there are no errors by doing `Kernel -> Restart Kernel and Run All Cells...`
2. Commit and push all your work (including the .ipynb file) to GitHub.
3. Submit a link to your repo on Canvas (this has to be done only once per group).
